## 넘파이 ndarray 개요

```python
import numpy as np
```

```python
#np.array() 사용: ndarray로 변환을 원하는 객체를 인자로 입력하면 ndarray를 반환함
#ndarray.shape는 ndarray의 차원과 크기를 튜플 형태로 나타냄
array1 = np.array([1, 2, 3])
print('array1 type:', type(array1))
print('array1 array 형태:', array1.shape)

array2 = np.array([[1, 2, 3],
                 [2, 3, 4]])
print('array2 type:', type(array2))
print('array2 array 형태:', array2.shape)

array3 = np.array([[1, 2, 3]])
print('array3 type:', type(array3))
print('array3 array 형태:', array3.shape)
```

```python
#각 array의 차원을 ndarray.dim을 이용해 확인
print('array1: {:0}차원, array2: {:1}차원, array3: {:2}차원'.format(array1.ndim, array2.ndim, array3.ndim))
```

```python
#ndarray의 데이터 타입
#한 개의 ndarray 객체에 int와 float가 함께 있을 수 없음
list1 = [1 ,2, 3]
print(type(list1))

array1 = np.array(list1)
print(type(array1))
print(array1, array1.dtype)
```

```python
#ndarray 내 데이터값의 타입 변경 astype() 메서드
#float->int 메모리 절약

#int32 --> float64 --> int32
array_int = np.array([1, 2, 3])
array_float = array_int.astype('float64')
print(array_float, array_float.dtype)

array_int1 = array_float.astype('int32')
print(array_int1, array_int1.dtype)

array_float1 = np.array([1.1, 2.1, 3.1])
array_int2 = array_float.astype('int32')
print(array_int2, array_int2.dtype) #소숫점 이하 모두 없어짐
```

```python
#ndarray 편리하게 생성하기: arange, zeros, ones

#arange(): array를 range()로 표현
sequence_array = np.arange(10)
print("sequence array:\n", sequence_array)
print(sequence_array.dtype, sequence_array.shape)
print("\n")

#zeros()
zero_array = np.zeros((3,2), dtype='int32')
print("zero array:\n", zero_array)
print(zero_array.dtype, zero_array.shape)
print("\n")

#ones()
one_array = np.ones((3,2))
print("one_array:\n", one_array)
print(one_array.dtype, one_array.shape)
print("\n")
```

```python
#ndarray의 차원과 크기를 변경하는 reshape()
print('1차원 array->2x5 array->5x2 array')

array1 = np.arange(10)
print('array1:\n', array1)

array2 = array1.reshape(2,5)
print('array2:\n', array2)

array3 = array1.reshape(5,2)
print('array3:\n', array3)
```

```python
#reshape에 인자로 -1을 적용하면 원래 ndarray와 호환되는 새로운 shape로 변환해줌
array1 = np.arange(10)
print(array1)
array2 = array1.reshape(-1, 5)
print('array2 shape:', array2.shape)
array3 = array1. reshape(5, -1)
print('array3 shape:', array3.shape)
```

```python
#reshape(1, -1)을 이용해 리스트 자료형으로 변환
array1 = np.arange(8) #1차원 array
array3d = array1.reshape((2, 2, 2)) #3차원 array
print('arrray3d:\n', array3d.tolist())

#3차원 ndarray를 2차원 ndarray로 변환
array5 = array3d.reshape(-1, 1)
print('array5:\n', array5.tolist())
print('array5 shape:', array5.shape)

#1차원 ndarray를 2차원 ndarray로 변환
array6 = array1.reshape(-1,1)
print('array6:\n', array6.tolist())
print('array6 shape:', array6.shape)
```

```python
#넘파이의 ndarray의 데이터 세트 확인하기-indexing

#단일 값 추출
#1부터 9까지의 1차원 ndarray 생성
array1 = np.arange(start=1, stop=10)
print('array1:', array1)

#index는 0부터 시작하므로 array1[2]는 3번째 index 위치의 데이터값을 의미
value = array1[2] #더 이상 ndarray 타입이 아니고 ndarray 내의 데이터값을 의미
print('value:', value)
print(type(value))
print('맨 뒤의 값:', array1[-1], '맨 뒤에서 두 번째 값:', array1[-2])
```

```python
array1[0]=9
array1[8]=0
print('array1:', array1)
```

```python
#다차원 ndarray에서 단일 값 추출
#1차원과 2차원의 차이: 2차원의 경우 콤마로 분리된 로우와 칼럼 위치의 인덱스를 통해 접근

array1d = np.arange(start=1, stop=10)
array2d = array1d.reshape(3,3)
print(array2d)

print('(row=0, col=0) index 가리키는 값:', array2d[0,0])
print('(row=0, col=1) index 가리키는 값:', array2d[0,1])
print('(row=1, col=0) index 가리키는 값:', array2d[1,0])
print('(row=2, col=2) index 가리키는 값:', array2d[2,2])

#이해를 돕기 위해 row와 column으로 표현했지만 사실은 axis0=0, axis1=1이 정확한 표현
#3차원 ndarray경우는 axis 0, axis 1, axis 2 3개의 축을 가지게 됨
#axis 0이 로우방향, axis 1이 칼럼 방향
```

```python
#슬라이싱
array1 = np.arange(start=1, stop=10)
array3 = array1[0:3] #0, 1, 2 index에 있는 데이터 반환
print(array3)
print(type(array3))
```

```python
array1 = np.arange(start=1, stop=10)
array4 = array1[:3]
print(array4)

array5 = array1[3:]
print(array5)

array6 = array1[1:]
print(array6)
```

```python
#2차원 슬라이싱
array1d = np.arange(start=1, stop=10)
array2d = array1d.reshape(3,3)
print('array2d:\n', array2d)

print('array2d[0:2, 0:2] \n', array2d[0:2, 0:2])
print('array2d[1:3, 0:3] \n', array2d[1:3, 0:3])
print('array2d[1:3, :] \n', array2d[1:3, :])
print('array2d[:, :] \n', array2d[:, :])
print('array2d[:2, 1:] \n', array2d[:2, 1:])
print('array2d[:2, 0] \n', array2d[:2, 0]) #칼럼 축 한쪽에만 슬라이싱 적용하고, 다른 쪽 축에는 단일 값 인덱스를 적용
```

```python
#array2d[0] --> 로우 축의 첫번째 로우 ndarray를 반환
#3차원 ndarray에서 뒤에 오는 인덱스를 없애면 2차원 ndarray 반환

print(array2d[0])
print(array2d[1])
print('array2d[0] shape:', array2d[0].shape, 'array2d[1] shape:', array2d[1].shape)
```

```python
**#팬시 인덱싱
#리스트나 ndarray로 인덱스 집합을 지정하면 해당 위치의 인덱스에 해당하는 ndarray를 반환하는 인덱싱 방식

array1d = np.arange(start=1, stop=10)
array2d = array1d.reshape(3,3)

array3 = array2d[[0,1], 2]
print('array2d[[0,1], 2] => ', array3.tolist())
#(row, col) 인덱스가 (0,2), (1,2)로 적용돼 [3,6]을 반환함

array4 = array2d[[0, 1], 0:2]
print('array2d[[0, 1], 2] => ', array4.tolist())
#((0,0), (0,1)), ((1,0), (1,1)) 인덱싱이 적용됨

array5 = array2d[[0, 1]]
print('array2d[0,1] => ', array5.tolist())
#((0, :), (1:0)) 인덱싱이 적용됨
```

```python
#불린 인덱싱: 조건 필터링과 검색을 동시에 할 수 있어 매우 자주 사용되는 인덱싱 방식
array1d = np.arange(start=1, stop=10)

#[] 안에 array1d > 5 Boolean indexing을 적용
array3 = array1d[array1d>5] #ndarray의 인덱스를 지정하는 [] 내에 조건문을 그대로 기재하기만 하면 됨
print('array1d>5 불린 인덱싱 결과 값:', array3)

'''
Step1. array1d>5와 같이 ndarray의 필터링 조건을 [] 안에 기재
Step2. False값은 무시하고 True값에 해당하는 인덱스값만 저장
(True값 자제가 아닌 True값을 가진 인덱스를 저장)
Step3. 저장된 인덱스 데이터 세트로 ndarray 조회
'''
```

```python
#numpy 객체에 조건식을 할당하면 어떤 일이 일어나느지 확인
array1d>5
#False, True로 이루어진 ndarray 객체 반환
```

```python
#행렬의 정렬 -sort() 와 argsort()
#np.sort(), ndarray.sort(), argsort()

#행렬 정렬
#np.sort()와 같이 넘파이에서 sort()를 호출. 원 행렬 그대로 유지한 채 원 행렬의 정렬된 행렬을 반환
#ndarray.sort()와 같이 행렬 자체에서 sort()를 호출. 원 행렬 자체를 정렬한 형태로 변환하며 반환 값은 None

org_array = np.array([3, 1, 9, 5])
print('원본 행렬:', org_array)

#np.sort()로 정렬
sort_array1 = np.sort(org_array)
print('np.sort() 호출 후 반환된 정렬 행렬:', sort_array1)
print('np.sort() 호출 후 원본 행렬:', org_array)

#ndarray.sort()로 정렬
sort_array2 = org_array.sort()
print('org_array.sort() 호출 후 반환된 행렬:', sort_array2)
print('org_array.sort() 호출 후 원본 행렬', org_array)
```

```python
#기본적으로 오름차순. 내림차순 적용하기 위해서는 np.sort()[::-1]

sort_array1_desc = np.sort(org_array)[::-1]
print('내림차순으로 정렬:', sort_array1_desc)
```

```python
#행렬이 2차원 이상일 경우

array2d = np.array([[8,12], 
                    [7,1]])
sort_array2d_axis0 = np.sort(array2d, axis=0) #로우 방향으로 정렬
print('로우 방향으로 정렬:\n', sort_array2d_axis0)

sort_array2d_axis1 = np.sort(array2d, axis=1)
print('칼럼 방향으로 정렬:\n', sort_array2d_axis1)
```

```python
#정렬된 행렬의 인덱스 반환하기: np.argsort()

org_array = np.array([3, 1, 9, 5])
sort_indices = np.argsort(org_array)
print(type(sort_indices))
print('행렬 정렬 시 원본 행렬의 인덱스:', sort_indices)
org_array = np.array([3, 1, 9, 5])
sort_indices_desc = np.argsort(org_array)[::-1]
print('행렬 내림차순 정렬 시 원본 행렬의 인덱스:', sort_indices_desc)
```

```python
'''
넘파이의 ndarray는 RDBMS의 TABLE칼럼이나 판다스 DataFrame 칼럼과 같은 메타 데이터를 가질 수 없음
실제 값과 그 값이 뜻하는 메타 데이터를 별도의 ndarray로 각각 가져야 함
'''
import numpy as np

name_array = np.array(['John', 'Mike', 'Sarah', 'Kate', 'Samuel'])
score_array = np.array([78, 95, 94, 98, 88])

sort_indices_asc = np.argsort(score_array) #정렬하기
print('성적 오름차순 정렬 시 score_array의 인덱스:', sort_indices_asc) #원본행렬의 인덱스
print('성적 오름차순으로 name_arry의 이름 출력:', name_array[sort_indices_asc]) #정렬 시 이름 출력
```

```python
#선형대수 연산 - 행렬 내적과 전치 행렬 구하기

#행렬 내적(행렬 곱): np.dot()

A = np.array([[1, 2, 3], [4, 5, 6]])
B = np.array([[7, 8], [9, 10], [11, 12]])
dot_product = np.dot(A, B)
print('행렬 내적 결과:\n', dot_product)
```

```python
#전치 행렬(Transpose): 원 행렬에서 행과 열의 위치를 교환한 원소로 구성한 행렬
A = np.array([[1,2], [3, 4]])
transpose_mat = np.transpose(A)
print('A의 전치행렬:\n', transpose_mat)
```

```python
#apply lambda식으로 데이터 가공
#판다스는 apply함수에 lambda 식을 결합해 DataFrame이나 Series의 레코드별로 데이터를 가공하는 기능을 제공합니다

#lambda는 파이썬에서 함수 프로그래밍을 지원하기 위하여 만들어짐

def get_square(a):
    return a**2

print('3의 제곱은:', get_square(3))
```

```python
#lambda로 표현

#:의 왼쪽에 있는 x는 입력 인자를 가리키며, 오른쪽은 입력 인자의 계산식
lambda_square = lambda x: x**2
print('3의 제곱은:', lambda_square(3))
```

```python
#lambda를 이용할 때 여러 개의 값을 입력 인자로 사용해야 할 경우에는 보통 map() 함수 결합
a=[1, 2, 3]
squares = map(lambda x: x**2, a)
list(squares)
```

```python
#Name 칼럼의 문자열 개수를 별도의 칼럼인 Name_len에 생성
titanic_df['Name_len']=titanic_df['Name'].apply(lambda x : len(x))
titanic_df[['Name', 'Name_len']].head(3)
```

```python
#lambda에서 if else 절을 사용해 좀 더 복잡한 가공을 함
#나아가 15세 미만이면 Child, 그렇지 않으면 Adult로 구분하는 새로운 칼럼 Child_Adult를 apply lambda를 이용해 만듦
titanic_df['Child_Adult']=titanic_df['Age'].apply(lambda x: 'Child' if x<=15 else 'Adult')
titanic_df[['Age', 'Child_Adult']].head(8)
```

```python
#나이가 15세 이하이면 Child, 15세=60세 사이는 Adult, 61세 이상은 Elderly로 분류하는 'Age_Cat' 칼럼을 만들어 보겠습니다.
titanic_df['Age_cat']=titanic_df['Age'].apply(lambda x: 'Child' if x<= 15 else ('Adult' if x<=60 else 'Elderly'))
titanic_df['Age_cat'].value_counts()
```

```python
#나이에 따라 세분화된 분류를 수행하는 함수 생성
def get_category(age):
    cat=''
    if age<=5: cat = 'Baby'
    elif age<= 12: cat = 'Child'
    elif age <= 18: cat = 'Teenager'
    elif age<= 25: cat = 'Student'
    elif age<=35: cat = 'Young Adult'
    elif age<=60: cat = 'Adult'
    else: cat = 'Elderly'
        
    return cat

#lambda 식에 위에서 생성한 get_category() 함수를 반환값으로 지정
#get_category(X)는 입력ㄱ밧으로 'Age' 칼럼 값을 받아서 해당하는 cat 반환
titanic_df['Age_cat']=titanic_df['Age'].apply(lambda x: get_category(x))
titanic_df[['Age', 'Age_cat']].head()
```
